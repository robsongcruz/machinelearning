## Análise 2

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df.info()
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
plt.figure(figsize=(12, 10))
ax = sns.countplot(y='brand', data=df)
ax.set(xlabel='Preço', ylabel='Marca', title='Preço Médio de Veículo por Marca')
```

```python
# Salvando o plot
fig = ax.get_figure()
fig.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='vehicleType', hue='gearbox', y='price', data=df)
ax.set(xlabel='Tipo de Veículo', ylabel='Preço', title='Preço Médio por Veículo e Marca')
```

```python
# Salvando o plot
fig = ax.get_figure()
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
