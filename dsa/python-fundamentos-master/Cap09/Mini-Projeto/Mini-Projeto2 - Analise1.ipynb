## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
```

```python
plt.figure(figsize=(12, 6))
ax = sns.distplot(df['yearOfRegistration'])
ax.set(xlabel='Ano de Registro', ylabel='Densidade (KDE)', title='Distribuição por Ano de Registro')
```

```python
# Salvando o plot
fig = ax.get_figure()
fig.savefig('plots/Analise1/vehicle-distribution.png')
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
plt.figure(figsize=(12, 6))
ax = sns.boxplot(x='vehicleType', y='price', data=df)
ax.set(xlabel='Tipo de Veículo', ylabel='Preço', title='Análise')
```

```python
# Salvando o plot
fig = ax.get_figure()
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='vehicleType', data=df)
ax.set(xlabel='Tipo de Veículo', ylabel='Unidades à Venda', title='Veículos à Venda')
```

```python
# Salvando o plot
fig = ax.get_figure()
fig.savefig("plots/Analise1/count-vehicleType.png")
```
