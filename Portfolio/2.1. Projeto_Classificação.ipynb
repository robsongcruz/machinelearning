```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score, train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
```

```python
url = 'data/sonar.all-data'
dataset = pd.read_csv(url, header=None)
```

#### 3. Preparar Dados
* a) Limpeza de dados
* b) Seleção de características

```python
X_train, X_validation, y_train, y_validation = train_test_split(dataset.drop(60, axis=1), dataset[60], test_size=0.2)
```

<a id="4."></a><br>
#### 4. Avaliar algoritmos
* a) Treinar algoritmos com o conjunto de treino
* b) Avaliar desempenho no conjunto de teste e/ou validação

- A avaliação inicial do conjunto de dados será realizada com vários algoritmos que são aplicáveis ao tipo de problema em questão utilizando a configuração default deles.

```python
class_models = []
class_models.append(('LR', LogisticRegressionCV()))
class_models.append(('LDA', LinearDiscriminantAnalysis()))
class_models.append(('KNN', KNeighborsClassifier()))
class_models.append(('CART', DecisionTreeClassifier()))
class_models.append(('NB', GaussianNB()))
class_models.append(('SVM', SVC()))
```

```python
num_folds = 10
scoring = 'accuracy'

results = []
names = []
for name, model in class_models:
    kfold = KFold(n_splits=num_folds)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = '%s: %f (%f)' % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# Comparação gráfica no desempenho dos Algoritmos
df_results = pd.DataFrame(results, index=['LR', 'LDA', 'KNN', 'CART', 'NB', 'SVM'])
df_results = df_results.transpose()
plt.title('Comparação gráfica no desempenho dos Algoritmos')
sns.boxplot(data=df_results, orient='h')
```

- Os parecem mais promissores para LR, LDA, KNN e SVM, porém os dois últimos são menos encorajadores já que apresentam alta variância comparado aos demais.
- Para avaliar de que forma as diferentes distribuições dos dados podem estar afetando negativamente os resultados, será realizado a verificado o desempenho com os dados padronizados.

```python
pipelines = []
pipelines.append(('ScaledLR', Pipeline([('Scaler', StandardScaler()),('LR', LogisticRegression())])))
pipelines.append(('ScaledLDA', Pipeline([('Scaler', StandardScaler()),('LDA', LinearDiscriminantAnalysis())])))
pipelines.append(('ScaledKNN', Pipeline([('Scaler', StandardScaler()),('KNN', KNeighborsClassifier())])))
pipelines.append(('ScaledCART', Pipeline([('Scaler', StandardScaler()),('CART', DecisionTreeClassifier())])))
pipelines.append(('ScaledNB', Pipeline([('Scaler', StandardScaler()),('NB', GaussianNB())])))
pipelines.append(('ScaledSVM', Pipeline([('Scaler', StandardScaler()),('SVM', SVC())])))

results = []
names = []
for name, model in pipelines:
    kfold = KFold(n_splits=num_folds)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# Comparação gráfica no desempenho dos Algoritmos com dados ajustados
df_results = pd.DataFrame(results, index=['Scaled LR', 'Scaled LDA', 'Scaled KNN', 'Scaled CART', 'Scaled NB', 'Scaled SVM'])
df_results = df_results.transpose()
plt.title('Comparação gráfica no desempenho dos Algoritmos (Scaled)')
sns.boxplot(data=df_results, orient='h')
```

- Quase todos os algoritmos tiveram a acurácia melhorada levemente com o ajuste, com certo prejuízo da variância. Os algoritmos que mais melhoraram foram CART e SVM.
