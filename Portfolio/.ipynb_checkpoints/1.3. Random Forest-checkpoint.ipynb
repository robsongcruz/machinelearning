#### 1. Preparação do problema
* a) Esclarecer o problema
* b) Carregar libraries
* c) Carregar dataset

- Os dados são originários do Lending Club, que conecta pessoas que precisam de dinheiro (mutuários) com pessoas que têm dinheiro (investidores). Como investidor, você gostaria de investir em pessoas que mostraram um perfil de ter uma alta probabilidade de pagá-lo de volta. Esses dados são de antes mesmo de serem públicos. Os dados são de empréstimos entre 2007 e 2010, a partir dos quais tentaremos classificar e prever se o mutuário pagou o empréstimo na íntegra.

As colunas da base de dados representam:

* credit.policy: 1 se o cliente atender aos critérios de subscrição de crédito da LendingClub.com e 0 caso contrário.
* purpose: O objetivo do empréstimo (leva valores "credit_card", "debt_consolidation", "educacional", "grande compra", "small_business" e "all_other").
* int.rate: a taxa de juros do empréstimo (uma taxa de 11% seria armazenada como 0,11). Os mutuários julgados por LendingClub.com para serem mais arriscados recebem taxas de juros mais elevadas.
* installment: as parcelas mensais devidas pelo mutuário se o empréstimo for financiado.
* log.annual.inc: O log natural da renda anual auto-relatada do mutuário.
* dti: Ratio dívida / rendimento do tomador do empréstimo (montante da dívida dividido pela receita anual).
* fico: a pontuação de crédito FICO do mutuário.
* days.with.cr.line: O número de dias em que o mutuário teve uma linha de crédito.
* revol.bal: Saldo rotativo do mutuário (montante não pago no final do ciclo de cobrança do cartão de crédito).
* revol.util: taxa de utilização da linha rotativa do mutuário (o valor da linha de crédito usada em relação ao crédito total disponível).
* inq.last.6mths: número de consultas do mutuário por credores nos últimos 6 meses.
* delinq.2yrs: o número de vezes que o mutuário havia passado mais de 30 dias em um pagamento nos últimos 2 anos.
* pub.rec: O número de registros públicos depreciativos do mutuário (arquivamentos de falências, ônus fiscais ou julgamentos).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score, train_test_split
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import plot_confusion_matrix, classification_report, confusion_matrix
%matplotlib inline
```

```python
# Sumarização e visualização dos dados
loans = pd.read_csv('data/loan_data.csv')
```

<a id="3."></a><br>
#### 3. Preparar Dados
* a) Limpeza de dados
* b) Seleção de características
* c) Transformação dos dados
* d) Dividir conjunto de dados em partes de treino e teste

```python
# Transformação de dados categóricos em numéricos
cat_feats = ['purpose']
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
final_data.head()
```

<a id="4."></a><br>
#### 4. Avaliar algoritmos
* a) Treinar algoritmos com o conjunto de treino
* b) Avaliar desempenho no conjunto de teste

```python
# Divisão dos em dados em conjuntos de treino e teste
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
#Treinamento de árvore de decisão
dtree = ExtraTreesClassifier()
dtree.fit(X_train, y_train)
```

```python
# Predição com dados novos
y_pred = dtree.predict(X_validation)

#Verificação de desempenho da Árvore
conf_tree = confusion_matrix(y_validation, y_pred)
rep_tree = classification_report(y_validation, y_pred)
```

```python
#Treinamento de Floresta Aleatória
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
#Predição com dados novos
y_pred = rf.predict(X_validation)
```

```python
rel_rf = classification_report(y_validation, y_pred)
confusion_rf = confusion_matrix(y_validation, y_pred)
```

```python
#Comparação de desempenho entre os algoritmos
print('Relatório de Classificação e Matriz de Confusão - Árvore de Decisão')
print(rep_tree)
print(conf_tree)
```

```python
print('Relatório de Classificação e Matriz de Confusão - Floresta Aleatória')
print(rel_rf)
print(confusion_rf)
```

<a id="5."></a><br>
#### 5. Melhorar Desempenho
* a) Verificar desempenho do(s) algoritmo(s) com os dados normalizados ou padronizados.
* b) Verificar desempenho do(s) algoritmo(s) em diferentes ajustes de parâmetros da técnica.
* c) Avaliar relevância de atributos x ruídos.

```python
param_grid = dict(n_estimators=[10, 20, 40, 70,  100, 200, 500])
num_folds = 10
scoring = 'accuracy'
kfold = KFold(n_splits=num_folds)

drop_params = ['credit.policy', 'inq.last.6mths', 'delinq.2yrs', 'pub.rec', 'purpose_credit_card', 
                                  'purpose_debt_consolidation', 'purpose_educational', 'purpose_home_improvement', 
                                  'purpose_major_purchase', 'purpose_small_business']
# Para tentar melhorar mais, vamos remover colunas menos relevantes
X_train_s = X_train.drop(drop_params, axis=1)
X_validation_s = X_validation.drop(drop_params, axis=1)
```

```python
grid = GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid, scoring=scoring, cv=kfold)
grid_result = grid.fit(X_train_s, y_train)
print('Best: %f using %s' % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('%f (%f) with: %r' % (mean, stdev, param))
```

- Apesar do resultado indicar 200 como melhor número de estimadores, será escolhido 70, pois o aumento na complexidade não vale acréscimo de desempenho obtido.

```python
# Preparação e validação do modelo final do modelo de Árvore com os parâmetros adequados
rf = RandomForestClassifier(n_estimators=70)
rf.fit(X_train_s, y_train)

# # Validação

y_pred = rf.predict(X_validation_s)
print(confusion_matrix(y_validation, y_pred))
print(classification_report(y_validation, y_pred))
```

- Não é possível afirmar que houve melhoria no modelo ao remover alguns parâmetros, mas certamente é obtido um modelo mais simples com bem menos parâmetros.
