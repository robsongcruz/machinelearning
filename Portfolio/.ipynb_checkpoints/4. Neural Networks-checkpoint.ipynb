### 4.1 Aplicação 1: Iris Dataset + Keras

__ETAPAS:__<br>

<a id="1.1"></a><br>
#### 1. Preparação do problema
* a) Esclarecer o problema
* b) Carregar libraries
* c) Carregar dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix
from keras.utils import np_utils
```

```python
iris = datasets.load_iris()
```

<a id="1.2"></a><br>
#### 2. Sumarizar Dados
* a) Estatística descritiva
* b) Visualização de dados e estrutura
* c) EDA

```python
# NA
```

<a id="1.3"></a><br>
#### 3. Preparar Dados
* a) Limpeza de dados
* b) Seleção de características
* c) Transformação dos dados
* d) Dividir conjunto de dados em partes de treino e teste/validação

```python
X = iris.data
y = iris.target

y_dummy = np_utils.to_categorical(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_dummy, test_size = 0.3, random_state = 0)
```

<a id="1.4"></a><br>
#### 4. Avaliar algoritmos
* a) Treinar algoritmos com o conjunto de treino
* b) Avaliar desempenho no conjunto de teste e/ou validação

```python
nr_model = Sequential()
nr_model.add(Dense(units = 5, input_dim = 4))
nr_model.add(Dense(units = 4))

# Softmax é comumente usado em saídas multiclasse
# inits = número de classes
nr_model.add(Dense(units = 3, activation = 'softmax'))

nr_model.summary()
```

```python
nr_model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
nr_model.fit(X_train, y_train, epochs = 1000, validation_data = (X_test, y_test))
```

```python
y_pred = nr_model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
y_test_matrix = [np.argmax(t) for t in y_test]
y_pred_matrix = [np.argmax(t) for t in y_pred]
conf = confusion_matrix(y_test_matrix, y_pred_matrix)
```

```python
sns.heatmap(conf.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=iris.target_names,
            yticklabels=iris.target_names,
            cmap='RdYlGn')
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.title('Confusion Matrix')
```

<a id="2.1"></a><br>
#### 1. Preparação do problema
* a) Esclarecer o problema
* b) Carregar libraries
* c) Carregar dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import np_utils
import numpy as np
from sklearn.metrics import confusion_matrix
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

<a id="2.2"></a><br>
#### 2. Sumarizar Dados
* a) Estatística descritiva
* b) Visualização de dados e estrutura
* c) EDA

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
plt.imshow(X_train[2], cmap = 'gray')
```

<a id="2.3"></a><br>
#### 3. Preparar Dados
* a) Limpeza de dados
* b) Seleção de características
* c) Transformação dos dados
* d) Dividir conjunto de dados em partes de treino e teste/validação

```python
X_train = X_train.reshape((len(X_train), np.prod(X_train.shape[1:])))
X_test = X_test.reshape((len(X_test), np.prod(X_test.shape[1:])))
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train /= 255
X_test /= 255
```

```python
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

<a id="2.4"></a><br>
#### 4. Avaliar algoritmos
* a) Treinar modelo(s) do algoritmo(s) com o conjunto de treino
* b) Avaliar desempenho no conjunto de teste e/ou validação

```python
#Preparação estrutural da rede neural
rn_model = Sequential()
rn_model.add(Dense(units = 64, activation = 'relu', input_dim = 784))
rn_model.add(Dropout(0.2))
rn_model.add(Dense(units = 64, activation = 'relu'))
rn_model.add(Dropout(0.2))
rn_model.add(Dense(units = 64, activation = 'relu'))
rn_model.add(Dropout(0.2))
rn_model.add(Dense(units = 10, activation = 'softmax'))
rn_model.summary()
```

```python
rn_model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
log = rn_model.fit(X_train, y_train, epochs = 20, validation_data = (X_test, y_test))
```

```python
log.history.keys()
plt.plot(log.history['val_loss'])
plt.plot(log.history['val_accuracy'])
plt.legend(['val_loss', 'val_accuracy'], loc='best')
```

```python
y_pred = rn_model.predict(X_test)
```

```python
y_test_matrix = [np.argmax(t) for t in y_test]
y_pred_matrix = [np.argmax(t) for t in y_pred]
conf = confusion_matrix(y_test_matrix, y_pred_matrix)
```

```python
sns.heatmap(conf.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=[i for i in range(10)],
            yticklabels=[i for i in range(10)],
            cmap='plasma')
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.title('Confusion Matrix')
```
